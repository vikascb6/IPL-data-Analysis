```python
# prompt: create a pie plot for the data df that the win percentage of the column winner

# Calculate win percentages for each team
winner_counts = df['winner'].value_counts()
total_matches = len(df)
win_percentages = (winner_counts / total_matches) * 100

# Create the pie plot
plt.figure(figsize=(10, 8))
plt.pie(win_percentages, labels=win_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Win Percentage of Teams')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
